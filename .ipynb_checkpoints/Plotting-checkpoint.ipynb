```python
import os
```

```python
import numpy
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',',dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt #make shortcut
```

```python
sample = headers[1]

plt.plot(data[:,1], label=sample)
plt.xlabel('simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig(F'{sample}.png') #save figure high rez
```

```python
sample = headers[1]
sample2 = headers[2]

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.xlabel('simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig('two_samples.png', dpi=300) #save figure high rez
```

```python
num_columns=len(headers)
print(headers)
print(F'The number of columns is {num_columns}')

for col in range(1,num_columns):
    print(F'The column number is {col}.')
    plt.plot(data[:,col], label=headers[col])
    
plt.xlabel('simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig('all_samples.png')
```

```python
num_columns=len(headers)

for col in range(1,num_columns):
    plt.figure()
    plt.plot(data[:,col], label=headers[col])
    plt.xlabel('simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png')
```

```python
plt.plot(data[:,0], data[:,1])
```
